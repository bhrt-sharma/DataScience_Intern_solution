# Problem 1

Build a classifier to predict the outcome of a new patient with high accuracy. Also, remember that as a data-scientist working on healthcare problems, your intent should also be to minimize the number of false-negatives.

```python
cor = df.corr()
plt.figure(figsize=(15,8))
sns.heatmap(cor,annot=True,cmap='plasma',linecolor='black')
```

#### Spliting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=7)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=None)
X_train_temp =  pca.fit_transform(X_train)
X_test_temp  =  pca.transform(X_test)
variance = pca.explained_variance_ratio_
```

```python
variance
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlim(0,33,1)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlim(0,6,1)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

###### it is clear that only 4  features are important

```python
pca = PCA(n_components=5)
X_train =  pca.fit_transform(X_train)
X_test  =  pca.transform(X_test) 
```

### Modelling

###### scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
len(X_test[0])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

##### Logistic Regression

```python

from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(max_iter=200, )
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print('LogisticRegression accuracy_score : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## 86 % acc

###### xgboost

```python
from xgboost import XGBClassifier

clf= XGBClassifier(learning_rate=0.1, n_estimators=7, max_depth=1,
                        min_child_weight=4, 
                         seed=27)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print('XGboost accuracy_score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
clf.score(X_train, y_train)

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

##### AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
clf= AdaBoostClassifier( n_estimators=10,
    learning_rate=0.15,)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print('AdaBoostClassifier accuracy_score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
clf.score(X_train, y_train)

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Problem 2

Build a regression model to predict the recurrence time for patients whose outcome is R.

```python
recurrent_df = df[df['Outcome']==1]   # since  R = 1
recurrent_df.shape
```

```python

X = recurrent_df.drop("Time",axis=1)         #N = 0 and R = 1
y = recurrent_df['Time']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15, random_state=7)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
pca = PCA(n_components=None)
X_train_temp =  pca.fit_transform(X_train)
X_test_temp  =  pca.transform(X_test)
variance = pca.explained_variance_ratio_
```

##### model

```python
from xgboost import XGBRegressor

reg = XGBRegressor(learning_rate=0.1, n_estimators=30, max_depth=2,
                        min_child_weight=4, 
                         seed=27)

reg.fit(X_train, y_train)

y_pred = reg.predict(X_test)
from sklearn import metrics

print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
